```python
import os
import docx
from docx.shared import Pt, Inches
from docx.enum.text import WD_ALIGN_PARAGRAPH
from google import genai
from google.colab import files

# 1. а¶”а¶¶а¶ңа·ҡ Gemini API Key а¶‘а¶ҡ а¶ёа·ҷа¶ӯа¶ұа¶§ а¶Үа¶ӯа·”а¶Ҫа¶ӯа·Ҡ а¶ҡа¶»а¶ұа·Ҡа¶ұ
GEMINI_API_KEY = "AQ.Ab8RN6KlmFSOizI7_Muy9MpboPy953gSypxx68dbROTz-XNiXw"

# 2. а¶”а¶¶а¶§ а¶…а·Җа·Ға·ҠвҖҚа¶ә а¶ҙа·ҠвҖҚа¶»а·Ға·Ҡа¶ұ а¶ҙа¶ӯа·ҠвҖҚа¶»а¶ә а¶ҙа·’а·…а·’а¶¶а¶і а·Җа·’а·ға·Ҡа¶ӯа¶»а¶ә (Prompt) а¶ёа·ҷа¶ӯа¶ұ а¶Ҫа·’а¶әа¶ұа·Ҡа¶ұ
# (а¶Ӣа¶Ҝа·Ҹа·„а¶»а¶«а¶әа¶ҡа·Ҡ а¶Ҫа·ҷа·ғ а¶Ӣа·ға·ға·Ҡ а¶ҙа·ҷа·… а¶·а·һа¶ӯа·’а¶ҡ а·Җа·’а¶Ҝа·ҠвҖҚа¶әа·Ҹа·Җ а¶°а·Ҹа¶»а·Ҹ а·Җа·’а¶Ҝа·ҠвҖҚа¶әа·”а¶ӯа¶ә а¶ҙа·ҠвҖҚа¶»а·Ға·Ҡа¶ұ а¶ҙа¶ӯа·ҠвҖҚа¶»а¶әа¶ҡа·Ҡ)
prompt = """а¶·а·һа¶ӯа·’а¶ҡ а¶»а·Ҹа·Ға·’ а·ға·„ а¶’а¶ҡа¶ҡ а·Җа¶»а·Ҡа¶ңа·“а¶ҡа¶»а¶«а¶ә а¶әа¶§а¶ӯа·ҡ а·ға¶»а¶Ҫ а¶ёа¶§а·Ҡа¶§а¶ёа·ҡ а·ға·’а¶§ а¶үа¶ӯа·Ҹ а·ға¶Ӯа¶ҡа·“а¶»а·Ҡа¶« а¶ёа¶§а·Ҡа¶§а¶ё а¶Ҝа¶ҡа·Ҡа·Җа·Ҹ а·Җа·’а·„а·’а¶Ҝа·“ а¶әа¶ұ MCQ а¶ңа·җа¶§а·…а·” 20а¶ҡа·Ҡ а·ға·‘а¶ё MCQ а¶‘а¶ҡа¶ҡа¶§а¶ё а·Җа·’а·Җа¶»а¶« 5а¶ҡа·Ҡ а¶¶а·җа¶ңа·’а¶ұа·Ҡ а¶Үа¶ӯа·’ а¶ҙа¶»а·’а¶Ҝа·’ а¶ҙа·’а·…а·’а¶ӯа·”а¶»а·” а·ға·„а·’а¶ӯа·Җ а¶Ҫа¶¶а·Ҹ а¶Ҝа·ҷа¶ұа·Ҡа¶ұ. (а¶ҡа·қа¶« а·Җа¶Ҫа¶§ а¶…а¶Ҝа·Ҹа¶Ҫа·Җ а¶ңа·җа¶§а·…а·” а¶әа·ңа¶Ҝа¶ұа·Ҡа¶ұ а¶‘а¶ҙа·Ҹ)

а¶үа·„а¶ӯа·’а¶ұа·Ҡ а·ға¶ҡа·ға¶ұ а·ға·’а¶әа·…а·”а¶ё MCQ а¶ҙа·ҠвҖҚа¶»а·Ға·Ҡа¶ұ а¶·а·һа¶ӯа·’а¶ҡ а·Җа·’а¶Ҝа·ҠвҖҚа¶әа·Ҹ а·ға¶ёа·ға·Ҡа¶® а·Җа·’а·Ӯа¶әа¶§ а¶…а¶Ҝа·Ҹа¶Ҫ а·Җа¶ұ а¶ҙа¶»а·’а¶Ҝа·’ вҒЈа·ға¶ҡа·ға¶ұа·Ҡа¶ұ. а·ға¶»а¶Ҫ а¶ёа¶§а·Ҡа¶§а¶ёа·ҡ а·ға·’а¶§ а¶үа¶ӯа·Ҹ а¶¶а·”а¶Ҝа·Ҡа¶°а·’а¶ёа¶ӯа·Ҡ а¶Ҝа¶»а·”а·Җа¶ұа·Ҡ а¶Ӣа¶Ҝа·ҷа·ға·Ҹ а¶үа·„а¶ӯ а¶ңа·җа¶§а·…а·” а·ға¶ҡа·ға¶ұа·Ҡа¶ұ
"""

def generate_and_download_paper():
    if GEMINI_API_KEY == "а¶”а¶¶а¶ңа·ҡ_Gemini_API_Key_а¶‘а¶ҡ_а¶ёа·ҷа¶ӯа¶ұа¶§_а¶Үа¶ӯа·”а¶Ҫа¶ӯа·Ҡ_а¶ҡа¶»а¶ұа·Ҡа¶ұ":
        print("а¶ҡа¶»а·”а¶«а·Ҹа¶ҡа¶» а¶ҙа·ҠвҖҚа¶»а¶®а¶ёа¶әа·ҷа¶ұа·Ҡ а¶”а¶¶а¶ңа·ҡ Gemini API Key а¶‘а¶ҡ а¶Үа¶ӯа·”а¶Ҫа¶ӯа·Ҡ а¶ҡа¶»а¶ұа·Ҡа¶ұ!")
        return

    print("Gemini а¶ёа¶ҹа·’а¶ұа·Ҡ а¶ҙа·ҠвҖҚа¶»а·Ға·Ҡа¶ұ а¶ҙа¶ӯа·ҠвҖҚа¶»а¶ә а·ға¶ҡа·ға¶ёа·’а¶ұа·Ҡ а¶ҙа·Җа¶ӯа·“. а¶ҡа¶»а·”а¶«а·Ҹа¶ҡа¶» а¶»а·җа¶іа·“ а·ға·’а¶§а·’а¶ұа·Ҡа¶ұ...")

    # Gemini а·ға·ҡа·Җа·Ҹа¶Ҝа·Ҹа¶әа¶ҡа¶әа·Ҹ а·ға¶ёа·Ҡа¶¶а¶ұа·Ҡа¶° а¶ҡа·’а¶»а·“а¶ё
    client = genai.Client(api_key=GEMINI_API_KEY)

    # Gemini 2.5 Flash а¶ёа·Ҹа¶Ҝа·’а¶Ҫа·’а¶ә а¶·а·Ҹа·Җа·’а¶ӯа¶әа·ҷа¶ұа·Ҡ а¶ҙа·ҠвҖҚа¶»а·Ға·Ҡа¶ұ а¶ҙа¶ӯа·ҠвҖҚа¶»а¶ә а¶Ҫа¶¶а·Ҹ а¶ңа·җа¶ұа·“а¶ё
    response = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=prompt,
    )

    exam_text = response.text
    filename = "Physics_Generated_Paper.docx"

    # --- Word Document а¶‘а¶ҡ а¶ұа·’а¶»а·Ҡа¶ёа·Ҹа¶«а¶ә а¶ҡа·’а¶»а·“а¶ёа·ҡ а¶ҡа·ңа¶§а·ғ ---
    doc = docx.Document()

    # а¶ҙа·’а¶§а·”а·Җа·ҡ а¶ёа·Ҹа¶әа·’а¶ёа·Ҡ а·ға¶ҡа·ға·Ҡ а¶ҡа·’а¶»а·“а¶ё (а¶…а¶ҹа¶Ҫа·Ҡ 1)
    for section in doc.sections:
        section.top_margin = Inches(1)
        section.bottom_margin = Inches(1)
        section.left_margin = Inches(1)
        section.right_margin = Inches(1)

    # Font а¶‘а¶ҡ Arial а·ға·„ Size а¶‘а¶ҡ 12pt а¶Ҫа·ҷа·ғ а·ға·җа¶ҡа·ға·“а¶ё (а·ға·’а¶Ӯа·„а¶Ҫ а¶әа·”а¶ұа·’а¶ҡа·қа¶©а·Ҡ а¶ұа·’а·Җа·җа¶»а¶Ҝа·’а·Җ а¶ҙа·ҷа¶ұа·Ҡа·Җа·“а¶ёа¶§)
    style = doc.styles['Normal']
    font = style.font
    font.name = 'Arial'
    font.size = Pt(12)

    # Gemini Text а¶‘а¶ҡ а¶ҙа·ҡа·…а·’а¶әа·ҷа¶ұа·Ҡ а¶ҙа·ҡа·…а·’а¶әа¶§ Word Document а¶‘а¶ҡа¶§ а¶‘а¶ҡа¶ӯа·” а¶ҡа·’а¶»а·“а¶ё
    lines = exam_text.split('\n')
    for line in lines:
        cleaned_line = line.strip()

        # Markdown Bold а·ға¶Ҫа¶ҡа·”а¶«а·” (**) а¶үа·Җа¶ӯа·Ҡ а¶ҡа·’а¶»а·“а¶ё а·„а·қ а¶ҙа·’а¶»а·’а·ға·’а¶Ҝа·” а¶ҡа·’а¶»а·“а¶ё
        cleaned_line = cleaned_line.replace("**", "")

        if cleaned_line == "":
            doc.add_paragraph()
        else:
            p = doc.add_paragraph(cleaned_line)

            # а¶ҙа·…а¶ёа·” а¶ҙа·ҡа·…а·’а¶ә а¶ёа·Ҹа¶ӯа·ҳа¶ҡа·Ҹа·Җ а¶Ҫа·ҷа·ғ а·ға¶Ҫа¶ҡа·Ҹ а¶ёа·җа¶Ҝа¶§ а¶ҡа¶» Bold а¶ҡа·’а¶»а·“а¶ё
            if p.text == lines[0].replace("**", "").strip():
                p.alignment = WD_ALIGN_PARAGRAPH.CENTER
                p.runs[0].bold = True
                p.runs[0].font.size = Pt(14)

    # File а¶‘а¶ҡ а·ға·ҡа·Җа·Ҡ а¶ҡа·’а¶»а·“а¶ё
    doc.save(filename)
    print(f"\n'{filename}' а·ға·Ҹа¶»а·Ҡа¶®а¶ҡа·Җ а¶ұа·’а¶»а·Ҡа¶ёа·Ҹа¶«а¶ә а¶ҡа¶»а¶ұ а¶Ҫа¶Ҝа·“!")

    # а·ға·ҳа¶ўа·”а·Җа¶ё а¶ҙа¶»а·’а¶ңа¶«а¶ҡа¶әа¶§ а·„а·қ а¶Ҝа·”а¶»а¶ҡа¶®а¶ұа¶әа¶§ Download а¶ҡа·’а¶»а·“а¶ё
    print("а¶”а¶¶а¶ңа·ҡ а¶Ӣа¶ҙа·Ҹа¶Ӯа¶ңа¶әа¶§ (Device) а¶¶а·Ҹа¶ңа¶ӯ а·Җа·ҷа¶ёа·’а¶ұа·Ҡ а¶ҙа·Җа¶ӯа·“...")
    files.download(filename)

# а·Җа·җа¶©а·ға¶§а·„а¶ұ а¶ҡа·ҠвҖҚа¶»а·’а¶әа·Ҹа¶ӯа·Ҡа¶ёа¶ҡ а¶ҡа·’а¶»а·“а¶ё
generate_and_download_paper()
```

```python
#@title **Gemini Physics Paper Generator Control Panel (With Progress Bar - Fixed)**
import os
import time
import docx
from docx.shared import Pt, Inches
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.oxml import OxmlElement
from docx.oxml.ns import qn
from google import genai
from google.colab import files
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- 1. а¶ҙа¶»а·’а·Ға·“а¶Ҫа¶ҡ а¶…а¶ӯа·”а¶»а·”а¶ёа·”а·„а·”а¶«а¶ӯ (UI Widgets) а·ға¶ҡа·ға·Ҡ а¶ҡа·’а¶»а·“а¶ё ---

api_key_input = widgets.Password(
    value='',
    placeholder='AIzaSy...',
    description='API Key:',
    layout=widgets.Layout(width='90%'),
    disabled=False
)

# Syntax Error а¶ёа¶ҹа·„а·җа¶»а·“а¶ё а·ға¶іа·„а·Ҹ default prompt а¶‘а¶ҡ а·ға¶»а¶Ҫа·Җ а¶ӯа¶ұа·’ а¶ҙа·ҡа·…а·’а¶әа¶ҡа·’а¶ұа·Ҡ а¶Ҫа¶¶а·Ҹ а¶Ҝа·“ а¶Үа¶ӯ
prompt_input = widgets.Textarea(
    value='а¶Ӣа·ға·ға·Ҡ а¶ҙа·ҷа·… а¶·а·һа¶ӯа·’а¶ҡ а·Җа·’а¶Ҝа·ҠвҖҚа¶әа·Ҹа·Җ (A/L Physics) а·Җа·’а·Ӯа¶ә а¶ұа·’а¶»а·Ҡа¶Ҝа·ҡа·Ға¶әа¶§ а¶…а¶ұа·”а·Җ "а¶°а·Ҹа¶»а·Ҹ а·Җа·’а¶Ҝа·ҠвҖҚа¶әа·”а¶ӯа¶ә" а¶ҙа·Ҹа¶©а¶ёа·ҡ а¶ҡа·’а¶»а·Ҡа¶ а·ңа·Ҷа·Ҡ а¶ұа·’а¶әа¶ё (Kirchhoff Laws) а¶ҙа¶Ҝа¶ұа¶ёа·Ҡ а¶ҡа¶»а¶ңа·ҷа¶ұ а¶»а·–а¶ҙ а·ға¶§а·„а¶ұа·Ҡ а¶Үа·ға·”а¶»а·’а¶ұа·Ҡ а·Җа·’а·ға¶іа·’а¶ә а¶әа·”а¶ӯа·” а¶¶а·„а·”а·Җа¶»а¶« а¶ҙа·ҠвҖҚа¶»а·Ға·Ҡа¶ұ (MCQ) 5а¶ҡа·Ҡ а·ға·„ а¶ҙа¶»а·’а¶ҙа¶® а·ға¶§а·„а¶ұа¶ҡа·Ҡ а·ға·„а·’а¶ӯ а·Җа·ҠвҖҚа¶әа·”а·„а¶ңа¶ӯ а¶»а¶ а¶ұа·Ҹ а¶ҙа·ҠвҖҚа¶»а·Ға·Ҡа¶ұа¶әа¶ҡа·Ҡ (Structured Essay) а·ға·’а¶Ӯа·„а¶Ҫ а¶ёа·Ҹа¶°а·ҠвҖҚа¶әа¶әа·ҷа¶ұа·Ҡ а·ға¶ҡа·ға·Ҡ а¶ҡа¶»а¶ұа·Ҡа¶ұ. а¶ҙа·’а·…а·’а¶ӯа·”а¶»а·” а¶ҙа¶ӯа·ҠвҖҚа¶»а¶ә а¶…а·Җа·ға·Ҹа¶ұа¶әа¶§ а¶Үа¶ӯа·”а¶Ҫа¶ӯа·Ҡ а¶ҡа¶»а¶ұа·Ҡа¶ұ. а¶ҡа·’а·ға·’а¶Ҝа·” а¶…а¶ёа¶ӯа¶» а·„а·җа¶іа·’а¶ұа·Ҡа·Җа·“а¶ёа¶ҡа·Ҡ а¶ұа·җа¶ӯа·’а·Җ а¶ҙа·ҠвҖҚа¶»а·Ға·Ҡа¶ұ а¶ҙа¶ӯа·ҠвҖҚа¶»а¶ә а¶ҙа¶ёа¶«а¶ҡа·Ҡ а¶Ҫа¶¶а·Ҹа¶Ҝа·ҷа¶ұа·Ҡа¶ұ.',
    placeholder='а¶”а¶¶а¶§ а¶…а·Җа·Ға·ҠвҖҚа¶ә а¶ҙа·ҠвҖҚа¶»а·Ға·Ҡа¶ұ а¶ҙа¶ӯа·ҠвҖҚа¶»а¶әа·ҡ а·Җа·’а·ға·Ҡа¶ӯа¶»а¶ә (Prompt) а¶ёа·ҷа¶ӯа¶ұ а¶Ҫа·’а¶әа¶ұа·Ҡа¶ұ...',
    description='Prompt:',
    layout=widgets.Layout(width='90%', height='150px'),
    disabled=False
)

download_plain_button = widgets.Button(
    description='Generate Plain Paper',
    disabled=False,
    button_style='info',
    tooltip='а¶»а·–а¶ҙ а·ға¶§а·„а¶ұа·Ҡ а¶»а·„а·’а¶ӯ а·ға·Ҹа¶ёа·Ҹа¶ұа·ҠвҖҚа¶ә а¶ҙа·ҠвҖҚа¶»а·Ға·Ҡа¶ұ а¶ҙа¶ӯа·ҠвҖҚа¶»а¶ә а¶Ҫа¶¶а·Ҹ а¶ңа·җа¶ұа·“а¶ёа¶§',
    icon='file-text'
)

download_diagram_button = widgets.Button(
    description='Generate & Download with Diagrams',
    disabled=False,
    button_style='success',
    tooltip='а¶»а·–а¶ҙ а·ға¶§а·„а¶ұа·Ҡ а·ға¶іа·„а·Ҹ а·Җа·ҷа¶ұа·Ҡа·Җа·– а¶»а·Ҹа¶ёа·” а·ға·„а·’а¶ӯ а¶Ҷа¶ҡа¶»а·Ҡа·Ӯа¶«а·“а¶ә а¶ҙа·ҠвҖҚа¶»а·Ға·Ҡа¶ұ а¶ҙа¶ӯа·ҠвҖҚа¶»а¶ә а¶Ҫа¶¶а·Ҹ а¶ңа·җа¶ұа·“а¶ёа¶§',
    icon='image'
)

progress_bar = widgets.IntProgress(
    value=0, min=0, max=100, step=1,
    description='Progress:',
    bar_style='info',
    orientation='horizontal',
    layout=widgets.Layout(width='90%', visibility='hidden')
)

status_label = widgets.Label(
    value='',
    layout=widgets.Layout(width='90%')
)

output_area = widgets.Output()

def set_cell_margins(cell, top=100, bottom=100, left=150, right=150):
    tcPr = cell._tc.get_or_add_tcPr()
    tcMar = OxmlElement('w:tcMar')
    for m, val in [('w:top', top), ('w:bottom', bottom), ('w:left', left), ('w:right', right)]:
        node = OxmlElement(m)
        node.set(qn('w:w'), str(val))
        node.set(qn('w:type'), 'dxa')
        tcMar.append(node)
    tcPr.append(tcMar)

# --- 2. а¶ҙа·ҠвҖҚа¶»а¶°а·Ҹа¶ұ а¶ҙа·ҠвҖҚа¶»а·Ға·Ҡа¶ұ а¶ҙа¶ӯа·ҠвҖҚа¶» а·ға·җа¶ҡа·ға·”а¶ёа·Ҡ а¶ҡа·ҠвҖҚа¶»а·’а¶әа·Ҹа·Җа¶Ҫа·’а¶ә ---
def generate_paper(with_diagrams=False):
    with output_area:
        clear_output()

        user_api_key = api_key_input.value.strip()
        if not user_api_key:
            status_label.value = "вқҢ Error: Please enter your Gemini API Key first!"
            return

        base_prompt = prompt_input.value.strip()
        if not base_prompt:
            status_label.value = "вқҢ Error: Please enter a prompt in the text box!"
            return

        # UI Progress Bar а¶‘а¶ҡ а¶ҙа·ҷа¶ұа·Ҡа·Җа·“а¶ё а¶Ҷа¶»а¶ёа·Ҡа¶· а¶ҡа·’а¶»а·“а¶ё
        progress_bar.layout.visibility = 'visible'
        progress_bar.value = 10
        progress_bar.bar_style = 'info'
        status_label.value = "вҸі Connecting to Gemini API..."

        final_prompt = base_prompt
        if with_diagrams:
            final_prompt += " (IMPORTANT: For questions that need circuits, graphs, or diagrams, explicitly add a line '[DIAGRAM: Describe what should be drawn here]' right after the question text, so I can format it properly.)"

        try:
            client = genai.Client(api_key=user_api_key)

            progress_bar.value = 40
            status_label.value = "вҸі Gemini is creating the exam paper content..."

            response = client.models.generate_content(
                model='gemini-2.5-flash',
                contents=final_prompt,
            )

            exam_text = str(response.text)
            filename = "Physics_Interactive_Paper.docx" if with_diagrams else "Physics_Plain_Paper.docx"

            progress_bar.value = 70
            status_label.value = "вҸі Writing text elements into Word document structure..."
            time.sleep(0.3)

            doc = docx.Document()

            for section in doc.sections:
                section.top_margin = Inches(1)
                section.bottom_margin = Inches(1)
                section.left_margin = Inches(1)
                section.right_margin = Inches(1)

            style = doc.styles['Normal']
            font = style.font
            font.name = 'Arial'
            font.size = Pt(12)

            lines = exam_text.split('\n')
            for index, line in enumerate(lines):
                cleaned_line = line.strip().replace("**", "").replace("###", "")

                if cleaned_line == "":
                    doc.add_paragraph()
                    continue

                if with_diagrams and ("[DIAGRAM:" in cleaned_line or "а¶»а·–а¶ҙ а·ға¶§а·„а¶ұ" in cleaned_line or "[DIAGRAM]" in cleaned_line):
                    table = doc.add_table(rows=1, cols=1)
                    table.alignment = docx.enum.table.WD_TABLE_ALIGNMENT.CENTER
                    cell = table.cell(0, 0)
                    cell.width = Inches(5.5)
                    set_cell_margins(cell, top=400, bottom=400, left=400, right=400)

                    p_box = cell.paragraphs[0]
                    p_box.alignment = WD_ALIGN_PARAGRAPH.CENTER
                    run_box = p_box.add_run(f"рҹ–јпёҸ [а¶»а·–а¶ҙ а·ға¶§а·„а¶ұ а·ға¶іа·„а·Ҹ а·Җа·ҷа¶ұа·Ҡ а¶ҡа·… а·ға·Ҡа¶®а·Ҹа¶ұа¶ә]\n({cleaned_line})")
                    run_box.font.name = 'Arial'
                    run_box.font.size = Pt(10)
                    run_box.italic = True
                    doc.add_paragraph()
                else:
                    p = doc.add_paragraph()
                    run = p.add_run(cleaned_line)

                    if index == 0:
                        p.alignment = WD_ALIGN_PARAGRAPH.CENTER
                        run.bold = True
                        run.font.size = Pt(14)

            progress_bar.value = 90
            status_label.value = "вҸі Completing download process..."
            doc.save(filename)
            time.sleep(0.3)

            progress_bar.value = 100
            progress_bar.bar_style = 'success'
            status_label.value = f"вң… Success: '{filename}' downloaded successfully!"

            files.download(filename)

        except Exception as e:
            progress_bar.value = 100
            progress_bar.bar_style = 'danger'
            status_label.value = f"вқҢ Process stopped: {repr(e)}"

def on_plain_clicked(b): generate_paper(with_diagrams=False)
def on_diagram_clicked(b): generate_paper(with_diagrams=True)

download_plain_button.on_click(on_plain_clicked)
download_diagram_button.on_click(on_diagram_clicked)

# --- 3. UI а¶‘а¶ҡ а¶ӯа·’а¶»а¶ә а¶ёа¶ӯ а¶ҙа·ҷа¶ұа·Ҡа·Җа·“а¶ё ---
print("вҡҷпёҸ Gemini Advanced Physics Paper Generator Control Panel:")
print("-" * 50)
display(api_key_input)
display(prompt_input)
print("-" * 50)
print("Select your preferred download mode:")
button_box = widgets.HBox([download_plain_button, download_diagram_button])
display(button_box)
print("-" * 50)
display(status_label)
display(progress_bar)
display(output_area)
```
